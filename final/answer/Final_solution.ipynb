```python
# problem 7
# iris datasets의 모든 column (SepalLength, SepalWidth, PetalLength, PetalWidth)
# 을 이용하여 decision tree를 만들고, 
# [4.8,3.1,1.5,0.2] 의 값을 예측하시오. (SK-Learn의 DecisionTreeClassifer 이용)

from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()


```

```python
# problem 7 solution

clf = tree.DecisionTreeClassifier(max_depth=5)
clf = clf.fit(iris.data, iris.target)
print(clf.predict([[4.8,3.1,1.5,0.2]]))
print(clf.predict_proba(([[4.8,3.1,1.5,0.2]])))
```

```python
# problem 9
# ratings 에서 movie 의 평균 평점이 4 이상인 영화의 비율을 구하시오. 
# (단, 영화 당 평점 데이터가 5개 이하인 경우는 제외한다.)

import pandas as pd
df = pd.read_csv("ratings.csv")
```

```python
# problem 9 solution

data = df.groupby('movieId').mean().rating
movie_count = df.groupby('movieId').size()
data_ov4 = data[movie_count[movie_count > 5 ].index]
print( data_ov4[data_ov4 >= 4].count()/data_ov4.count())
```
