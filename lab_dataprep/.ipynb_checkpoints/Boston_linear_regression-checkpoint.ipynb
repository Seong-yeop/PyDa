```python
from sklearn import datasets, linear_model
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('BostonHousing.csv')
#df = pd.DataFrame(boston, columns=boston.feature_names)
```

```python
df['LSTAT_log'] = preprocessing.scale(np.log(df['LSTAT']+1))
df['LSTAT_sqrt'] = preprocessing.scale(np.sqrt(df['LSTAT']+1))

plt.hist(df['LSTAT'], bins=20)
plt.show()

plt.hist(df['LSTAT_log'], bins=20)
plt.show()

plt.hist(df['LSTAT_sqrt'], bins=20)
plt.show()
```

```python
# log transformation
x = df.LSTAT.values.reshape(df.shape[0],1) 
x_log = df.LSTAT_log.values.reshape(df.shape[0],1)
y = df.MEDV.values.reshape(df.shape[0],1)
```

```python
regr = linear_model.LinearRegression()
regr.fit(x_log,y)
```

```python
plt.scatter(x_log,y)
plt.plot(x_log, regr.predict(x_log), 'b-')
plt.show()
```

```python
print(np.mean((regr.predict(x_log) - y)**2))
print(regr.score(x_log,y))
```

```python
# default 
regr = linear_model.LinearRegression()
regr.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x, regr.predict(x), 'b-')
plt.show()
```

```python
print(np.mean((regr.predict(x) - y)**2))
print(regr.score(x,y))
```
