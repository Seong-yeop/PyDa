```python
#### numpy, pandas basic ####

import numpy as np
import pandas as pd

def load_data(path):
    data = pd.read_csv("ratings.csv")
    return data

def get_ratings(df):
    return


path = 'ratings.csv'
m_id = 'movieId'

df = load_data(path)


arr = np.array(df)      
    
means = {}
for i in np.unique(arr[:,1]):
    tmp = arr[np.where(arr[:,1] == i)]
    means[i] = np.mean(tmp[:,2])
for i in means:
    if(means[i] > 4.5):
        break
        #print(" : %d,  : %.2f"%(i,means[i]))

#movie_rank = {k: v for k, v in sorted(means.items(), key=lambda item: item[1])}

sorted_movies = sorted(means.items(), key=lambda kv: kv[1])
print(sorted_movies[-1:-100:-1])


rating = df.groupby(m_id).mean()
rating_over_4 = rating[rating.loc[:,'rating'] > 4.5]

    
df_movie = pd.read_csv("movies.csv")
    
df_merge = df.merge(df_movie, on=m_id)
merge_rating = df_merge.groupby('title').mean()
count = df_merge.movieId.value_counts()

count_dict = dict(count)
print(count_dict)


merge_rating_over_4 = merge_rating[merge_rating.loc[:,'rating'] > 4.5]
ranking = merge_rating_over_4.sort_values(by='movieId', ascending=False)

```
