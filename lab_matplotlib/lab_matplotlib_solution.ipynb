```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import seaborn as sns
```

```python
# lab 1 
df = pd.read_csv('sales_data.csv')
```

```python
plt.plot(df.month_number, df.facecream, label='facecream', marker='^' , linewidth=2)
plt.plot(df.month_number, df.facewash, label='facewash', marker='o' , linewidth=2)
plt.plot(df.month_number, df.toothpaste, label='toothpaste', marker='o' , linewidth=2)
plt.plot(df.month_number, df.bathingsoap, label='bathingsoap', marker='o' , linewidth=2)
plt.plot(df.month_number, df.shampoo, label='shampoo', marker='o' , linewidth=2)
plt.plot(df.month_number, df.moisturizer, label='moisturizer', marker='o' , linewidth=2)

plt.xlabel('month number')
plt.ylabel('sales units in number')
plt.xticks(df.month_number)
plt.yticks([i for i in range(0,15000,1000)])
plt.legend(loc='upper right')
plt.title('sales data')
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
# lab 3
ax = plt.subplots()
ax = sns.countplot(x="survived", data=titanic)
plt.title("number of survived or ...")

#plt.savefig('survived')
plt.show()
```

```python
# lab 4
# seaborn 으로 구현
sns.countplot(x='class', data=titanic) # seaborn 사용하는 방법 
plt.show()
```

```python
# lab 5
# matplotlib 으로 구현
class_data = titanic['class'].value_counts()
plt.bar(class_data.index, class_data)
#plt.savefig('1-4.jpg')
plt.xlabel('class')
plt.ylabel('number of people')
plt.title('number of people by class')
plt.savefig('lab4.jpg')
plt.show()
```

```python
# lab 6

# seaborn countplot version
sns.countplot(x="class", hue="sex", data=titanic)
plt.savefig('1')
plt.show()

# seaborn bar version
sns.barplot(x='class', hue='sex',y='survived' ,data=titanic, estimator=len)
plt.show()
```

```python
# lab 7
sns.barplot(x="class", y="survived", hue="sex", data=titanic)
#sns.countplot(x="class", hue="survived", data=titanic)
plt.title("survived percent by class, sex")
#plt.savefig('lab8')
plt.show()
```

```python
# lab 8
f, ax = plt.subplots(1,2,figsize=(12,8))
sns.violinplot(x='pclass', y='age', hue='survived',scale='count',split=True,data=titanic,ax=ax[0])
ax[0].set_yticks(range(0,110,10))

sns.violinplot(x='sex', y='age', hue='survived',scale='count',split=True, data=titanic, ax=ax[1])
ax[1].set_yticks(range(0,110,10))

plt.tight_layout()
plt.show()
```

```python
# lab 9

cols = ['survived', 'sex', 'pclass', 'sibsp', 'parch', 'embarked']
nr_rows = 2
nr_cols = 3

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*3.5,nr_rows*3))

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        i = r*nr_cols+c       
        ax = axs[r][c]
        #sns.countplot(titanic[cols[i]], hue=titanic["survived"], ax=ax)
        sns.countplot(x=cols[i], hue='survived', ax=ax, data=titanic)
        ax.set_title(cols[i], fontsize=14, fontweight='bold')
        ax.legend(title="survived", loc='upper center') 

plt.tight_layout()  
plt.show()
```
